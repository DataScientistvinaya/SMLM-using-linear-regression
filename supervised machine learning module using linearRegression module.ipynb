```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df['sex']=df['sex'].apply({'female':1,'male':0}.get)# converting catergorical data into numerical with onehot encoding
```

```python
df['smoker']=df['smoker'].apply({'yes':1,'no':0}.get)
```

```python
df['region']=df['region'].apply({'southwest':1,'southeast':0,'northwest':2,'northeast':3}.get)
```

```python
df.head()
```

```python
x=df[['age','sex','bmi','children','smoker','region',]]# divind the data into dependent and independent variable
```

```python
y=df[['charges']]
```

# Spliting the data set training and testing the data set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)# 80% of data is used from training
```

```python
#creating our SMLM using Linear Regression algorithim

from sklearn.linear_model import LinearRegression
regression=LinearRegression()# activating the algorithim
```

```python
regression.fit(x_train,y_train)#.fit() used to train
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
#accuracy of SMLM
regression.score(x,y)
```
